```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import io
import requests
```

```python
url = 'https://github.com/1010code/iris-dnn-tensorflow/raw/master/data/Iris.csv'
s = requests.get(url).content
df_train = pd.read_csv(io.StringIO(s.decode('utf-8')))
df_train = df_train.drop(labels=['Id'],axis = 1)
df_train
```

```python
label_map = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
df_train['label'] = df_train['Species'].map(label_map)
df_train
```

# 檢查缺失值

```python
#移除缺失值
df_train = df_train.dropna()
```

```python
x = df_train.drop(labels = ['Species','label'],axis = 1).values
print('checked missing data(nan count):' ,len(np.where(np.isnan(x))[0]))
```

# 資料前處理

```python
import numpy as np
np.set_printoptions(suppress= True)
```

```python
#standardization
from  sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(x)
x_scaled = scaler.transform(x)

print('資料x平均值:',x.mean(axis = 0))
print('資料x標準差:',x.std(axis=0))

print('\nStandardScaler 縮放過後的資料及x平均值:',x_scaled.mean(axis = 0))
print('\nStandardScaler 縮放過後的資料及x標準差:',x_scaled.std(axis = 0))
```

```python
fig, axes = plt.subplots(nrows = 1, ncols =4)
fig.set_size_inches(15,5)
sns.distplot(x_scaled[:,0],ax = axes[0])
sns.distplot(x_scaled[:,1],ax = axes[1])
sns.distplot(x_scaled[:,2],ax = axes[2])
sns.distplot(x_scaled[:,3],ax = axes[3])

axes[0].set(xlabel = 'sepalLengthCm', title = 'distribution of sepallengthCm')
axes[1].set(xlabel = 'sepalwidthhCm', title = 'distribution of sepalwidthhCm')
axes[2].set(xlabel = 'petalLengthCm', title = 'distribution of petallengthCm')
axes[3].set(xlabel = 'petalwidthhCm', title = 'distribution of petallengthCm')
```

```python
#minmaxscaler
from  sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1)).fit(x)
x_scaled = scaler.transform(x)

print('資料x平均值:',x.mean(axis = 0))
print('資料x標準差:',x.std(axis=0))

print('\nStandardScaler 縮放過後的資料及x平均值:',x_scaled.mean(axis = 0))
print('\nStandardScaler 縮放過後的資料及x標準差:',x_scaled.std(axis = 0))
```

```python
#maxabsScaler
from  sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler().fit(x)
x_scaled = scaler.transform(x)

print('資料x平均值:',x.mean(axis = 0))
print('資料x標準差:',x.std(axis=0))

print('\nStandardScaler 縮放過後的資料及x平均值:',x_scaled.mean(axis = 0))
print('\nStandardScaler 縮放過後的資料及x標準差:',x_scaled.std(axis = 0))
```

```python
#robustScaler
from  sklearn.preprocessing import RobustScaler
scaler = RobustScaler().fit(x)
x_scaled = scaler.transform(x)

print('資料x平均值:',x.mean(axis = 0))
print('資料x標準差:',x.std(axis=0))

print('\nStandardScaler 縮放過後的資料及x平均值:',x_scaled.mean(axis = 0))
print('\nStandardScaler 縮放過後的資料及x標準差:',x_scaled.std(axis = 0))
```
